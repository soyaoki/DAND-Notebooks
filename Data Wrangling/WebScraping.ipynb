```python
url = 'https://www.rottentomatoes.com/m/et_the_extraterrestrial'
respones = requests.get(url)
respones
```

```python
# save HTML to file
with open("et_the_extraterrestrial.html", mode='wb') as file:
    file.write(respones.content)
```
